```python
# Cell 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Set style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
%matplotlib inline

print("âš¡ Electric Motor Temperature Prediction - Day 1")
print("="*60)

# Cell 2: Load data
df = pd.read_csv('data/raw/motor_temperature_dataset.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])

print(f"âœ… Dataset loaded successfully!")
print(f"ğŸ“Š Shape: {df.shape[0]} rows, {df.shape[1]} columns")
print(f"ğŸ“… Date range: {df['timestamp'].min()} to {df['timestamp'].max()}")

# Cell 3: Basic information
print("\nğŸ“‹ Dataset Information:")
df.info()

print("\nğŸ“ˆ First 5 rows:")
df.head()

# Cell 4: Missing values
print("\nğŸ” Missing Values:")
missing = df.isnull().sum()
if missing.sum() == 0:
    print("âœ… No missing values!")
else:
    print(missing[missing > 0])

# Cell 5: Statistical summary
print("\nğŸ“Š Statistical Summary:")
df.describe()

# Cell 6: Target variable analysis
print("\nğŸ¯ Target Variable: Motor Temperature")
print(f"   Mean: {df['motor_temperature'].mean():.1f}Â°C")
print(f"   Std:  {df['motor_temperature'].std():.1f}Â°C")
print(f"   Min:  {df['motor_temperature'].min():.1f}Â°C")
print(f"   Max:  {df['motor_temperature'].max():.1f}Â°C")

# Plot distribution
plt.figure(figsize=(10, 6))
plt.hist(df['motor_temperature'], bins=50, edgecolor='black', alpha=0.7)
plt.axvline(df['motor_temperature'].mean(), color='red', linestyle='--', 
           label=f'Mean: {df["motor_temperature"].mean():.1f}Â°C')
plt.axvline(80, color='orange', linestyle='--', label='Warning Threshold (80Â°C)')
plt.axvline(100, color='red', linestyle='--', label='Critical Threshold (100Â°C)')
plt.xlabel('Motor Temperature (Â°C)')
plt.ylabel('Frequency')
plt.title('Distribution of Motor Temperature')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Cell 7: Correlation analysis
print("\nğŸ”— Correlation with Motor Temperature:")
correlations = df.corr()['motor_temperature'].sort_values(ascending=False)
print(correlations)

# Heatmap
plt.figure(figsize=(12, 10))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()

# Cell 8: Feature vs Temperature scatter plots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
features = [
    ('current', 'Current (A)'),
    ('rpm', 'RPM'),
    ('load_percentage', 'Load (%)'),
    ('ambient', 'Ambient Temp (Â°C)'),
    ('torque', 'Torque (Nm)'),
    ('electrical_power', 'Power (W)')
]

for idx, (col, title) in enumerate(features):
    ax = axes[idx//3, idx%3]
    ax.scatter(df[col], df['motor_temperature'], alpha=0.5, s=10)
    ax.set_xlabel(title)
    ax.set_ylabel('Motor Temperature (Â°C)')
    ax.set_title(f'{title} vs Temperature')
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Cell 9: Time series analysis
plt.figure(figsize=(15, 8))
plt.subplot(2, 1, 1)
plt.plot(df['timestamp'][:200], df['motor_temperature'][:200], 'b-', linewidth=2)
plt.xlabel('Time')
plt.ylabel('Temperature (Â°C)')
plt.title('Motor Temperature Over Time (First 200 samples)')
plt.grid(True, alpha=0.3)

plt.subplot(2, 1, 2)
plt.plot(df['timestamp'][:200], df['current'][:200], 'r-', linewidth=2)
plt.xlabel('Time')
plt.ylabel('Current (A)')
plt.title('Motor Current Over Time')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Cell 10: Save insights
print("\nğŸ’¡ Key Insights from Day 1:")
print("="*50)
print("1. âœ… Data loaded successfully - 10,000 samples")
print("2. âœ… No missing values - good data quality")
print("3. ğŸ”¥ Temperature range: {:.1f}Â°C to {:.1f}Â°C".format(
    df['motor_temperature'].min(), df['motor_temperature'].max()))
print("4. ğŸ”— Strongest correlations with temperature:")
for col, corr in correlations.head(5).items():
    print(f"   - {col}: {corr:.3f}")
print("5. â° Data spans multiple days with time series pattern")
print("6. âš ï¸ Some motors may exceed warning threshold (80Â°C)")
print("\nâœ… DAY 1 COMPLETE! Ready for preprocessing and modeling.")
```
